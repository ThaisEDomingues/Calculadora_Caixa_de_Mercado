```python
# sem usar numpy
#matematica basica

numero1 = float(input('informe o primeiro numero: '))
numero2 = float(input('informe o segundo numero: '))


# soma
soma = numero1 + numero2
print(soma)
# subtração
subtracao = numero1 - numero2
print(subtracao)
# divisão
divi = numero1/numero2
print(divi)
#multiplicação
multi = numero1 * numero2
print(multi)
```

```python
#usando biblioteca numpy
#matematica basica

numero1 = float(input('informe o primeiro numero: '))
numero2 = float(input('informe o segundo numero: '))

#soma pode ser sum ou add
print(f'soma: {np.add(numero1, numero2)}')

#o metodo sum- so para numeros inteirops
#print(f'soma: {np.sum(numero1, numero2)}')

#subtrair subtract

print(f'subtracao: {np.subtract(numero1, numero2)}')

#multiplicar multiply

print(f'multiplicacao: {np.multiply(numero1, numero2)}')

#divisão divide

print(f'divisao: {np.divide(numero1, numero2)}')
```

```python
#caixa de supermercado

#temos produtos
#temos quantidade
#temos um total

#declarando variaveis
lista_nome =  []
lista_valor = []
lista_quantidade = []

while True:
  op_tem = input('Tem algum produto para passar? S/N')
  if op_tem.lower()  == 'n':
     print('sistema finalizado')
     break
  elif op_tem.lower() == 's':
     while True:

#funcao que adiciona o valor na lista produtp, valor e quantidade
#nome-da-lista.append(indice)
        nome = input('qual é o produto? ')
        valor = float(input('qual é o valor? '))
        quantidade = float(input('qual é a quantidade? '))
        lista_nome.append(nome)
        lista_valor.append(valor)
        lista_quantidade.append(quantidade)
        op_novo_produto = input('Tem outro produto para passar? S/N')
        if op_novo_produto.lower() == 'n':
          print('indo para nota fiscal')
          print('-------CUPOM FISCAL-------')
          print('Produto | Valor | Quantidade | Total')
#no comando for para iteração das listas
#i index para produto
#j index para valor
#k index para quantidade
#l index para total
#O total é uma multiplicação de valor e quantidade
#comando zip(lista, lista ,lista) para colocar varias listas dentro do comando for

#transformar a lista em numpy
#np.array() - transforma um conjunto de numeros para numpy
          np_valor = np.array(lista_valor)
          np_quantidade = np.array(lista_quantidade)
        # np.array() - transforma um conjunto de numeros para numpy
          total = np.multiply(np_valor, np_quantidade)
          for i,j,k,l in zip(lista_nome, lista_valor, lista_quantidade,total):
           print(f'{i}--{j}--{k}--{l}')
          break
        else:
         print('informe o novo produto')
```
